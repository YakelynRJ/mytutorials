```python
# modified from https://github.com/isgiddy/giddy-2020-roammiz.git
def calc_wstress(u10_wind,v10_wind):

    import airsea
    tau_x=airsea.windstress.stress(u10_wind,z=10,drag='largepond',rho_air=1.22,Ta=10.)
    tau_y=airsea.windstress.stress(v10_wind,z=10,drag='largepond',rho_air=1.22,Ta=10.)

    tx = np.where(u10_wind>0,tau_x,tau_x*-1)  # Add directional signs back to wind stress vectors
    ty = np.where(v10_wind>0,tau_y,tau_y*-1)

    tau = np.sqrt(tx**2+ty**2)

    [tau2,theta]=cart2pol(u10_wind,v10_wind)   #to get winds oriented 0 for Easterly. 
    wind_dir=np.rad2deg(theta)

    return tau, tx, ty, wind_dir
# modified from https://github.com/rabernat/intro_to_physical_oceanography.git
def winds2ekman(taux,tauy,lat,lat_fill_nan):
    g        = 9.81          # Gravitational acceleration (m/s2)
    omega    =  7.292116E-5 # Rotation rate of earth (rad/s) 
    rho_ref  = 1025.   # Reference density for sea water (kg/m3)
    cp       = 3985         # Specific heat capacity of sea water (J/kg/K)
    f    = 2 * omega * np.sin(np.deg2rad(lat))
    uek  = tauy/(rho_ref * f)#.where(abs(lat>5))
    vek  = -1*taux/(rho_ref * f)#.where(abs(lat>5))
    mag = np.sqrt(uek*uek + vek*vek)
    ele = np.where(abs(lat[:,0])<=lat_fill_nan)[0]
    uek[ele,:] = np.nan
    vek[ele,:] = np.nan
    mag[ele,:] = np.nan
    return uek, vek, mag
```
